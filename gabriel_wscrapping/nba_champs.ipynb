```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
import pandas as pd
```

```python
service = Service(executable_path='../chrome/chromedriver.exe')
options = webdriver.ChromeOptions()
```

```python
driver = webdriver.Chrome(service=service, options=options)
```

```python
driver.get("https://es.wikipedia.org/wiki/Anexo:Campeones_de_la_NBA")
```

```python
element = driver.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[5]')
element
```

```python
rows = element.find_elements(By.TAG_NAME, 'tr')
rows
```

```python
team = []
championships = []
years = []

for row in rows:
    
    cells = row.find_elements(By.TAG_NAME, 'td')
    
    if not cells:
        cells = row.find_elements(By.TAG_NAME, 'th')
        
    cells_text = [cell.text for cell in cells]
    team.append(cells_text[0])
    championships.append(cells_text[1])
    years.append(cells_text[2])
```

```python
team
```

```python
dic = {'Teams': team[1:],
       'Championships': championships[1:],
       'Years': years[1:]}

dic
```

```python
df_NBA = pd.DataFrame(dic)
df_NBA
```

```python
df_NBA.index = pd.RangeIndex(start=1, stop=len(df_NBA) + 1, step=1)
```
